```python
# Dependencies
import pandas as pd
```

```python
# Read in raw data
file = "Resources/NFL Arrest CSV.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head(10) #956, 8
```

```python
# Makes all crimes lowercase, removes players listed as "Free agent" team from data
df.CATEGORY = df.CATEGORY.str.lower()
df = df[df.TEAM != "Free Agent"]
df = df[df.TEAM != "Free agent"]
df.TEAM.unique()
```

```python
# Displays players with multiple crimes in different columns
df[["Crime 1", "Crime 2", "Crime 3"]] = df.CATEGORY.str.split(",", expand=True)
df["crime_list"] = df.CATEGORY.str.split(",")
df["crime_count"] = df.crime_list.str.len()
df.head()
```

```python
# Renames columns for readability and usability
df_col_names = ['date', 'team', 'player', 'position', 'status', 'crimeraw', 
                'desc', 'outcome', 'crime1', 'crime2', 'crime3', 'crime_list', 'crime_count']
df.columns = df_col_names
clean_df = df
clean_df.head()
```

```python
# Exports dataframe to a CSV
clean_df.to_csv('Resources/clean_separated_list_count.csv', index=False)
```
