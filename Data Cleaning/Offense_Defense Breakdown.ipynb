```python
# Dependencies
import pandas as pd
```

```python
# Read in file
file = 'Resources/clean_separated_list_count.csv'
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head()
```

```python
offense = ['WR', 'RB', 'OT', 'TE', 'QB', 'OC', 'OG', 'FB', 'K', 'OL', 'C', 'P']
defense = ['CB', 'S', 'DT', 'LB', 'DE', 'DB', 'DT']
```

```python
offense_crime_df = df.loc[df.position.isin(offense), ["player", "position", "crimeraw", "crime_count"]]
offense_crime_df
```

```python
defense_crime_df = df.loc[df.position.isin(defense), ["player", "position", "crimeraw", "crime_count"]]
defense_crime_df
```

```python
# Exports the dataframes to CSV files
offense_crime_df.to_csv('Resources/offense_crime.csv', index=False)
defense_crime_df.to_csv('Resources/defense_crime.csv', index=False)
```
